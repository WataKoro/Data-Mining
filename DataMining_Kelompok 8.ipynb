# Merging dataframe peminat upi dengan dataframe jurusan

```python
dfupifstmajor = dfupifst.drop(['id_second_major', 'id_second_university'],axis=1)
dfupifstmajor = dfupifstmajor.rename(columns={'id_first_major': 'id_major', 'id_first_university': 'id_university'})
dfupifstmajor.head()
```

```python
dfmajor.head()
```

```python
dfupifstmajorfix = pd.merge(dfupifstmajor, dfmajor, how='left',on=["id_major"])
dfupifstmajorfix = dfupifstmajorfix.drop(['id_university_y', 'type', 'id_university_x', 'Unnamed: 0',
                                         'university_name', 'tka_score', 'tps_score'],axis=1)
dfupifstmajorfix.head()
```

# Membuat dataframe FPMIPA

```python
dfFPMIPA = dfupifstmajorfix.copy()
dfFPMIPA.drop(dfFPMIPA[~(dfFPMIPA.major_name == "PENDIDIKAN ILMU KOMPUTER") & ~(dfFPMIPA.major_name == "ILMU KOMPUTER") & ~(dfFPMIPA.major_name == "PENDIDIKAN MATEMATIKA") & ~(dfFPMIPA.major_name == "MATEMATIKA") & ~(dfFPMIPA.major_name == "PENDIDIKAN BIOLOGI") & ~(dfFPMIPA.major_name == "BIOLOGI") & ~(dfFPMIPA.major_name == "PENDIDIKAN FISIKA") & ~(dfFPMIPA.major_name == "FISIKA") & ~(dfFPMIPA.major_name == "PENDIDIKAN KIMIA") & ~(dfFPMIPA.major_name == "KIMIA") & ~(dfFPMIPA.major_name == "PENDIDIKAN ILMU PENGETAHUAN ALAM")].index, inplace = True)
```

# Persebaran skor matematika

```python
dfFPMIPA.groupby(["major_name"])["score_mat"].mean()
```

Analisis : Pendidikan Matematika dan Matematika memiliki rata-rata skor matematika lebih tinggi daripada jurusan lain, sementara Biologi memiliki rata-rata terendah. Sesuai karena Biologi cenderung lebih sedikit menggunakan matematika dibanding jurusan lain.
